```python
# Do NOT modify this code cell

# importing useful libraries
import re

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Do NOT modify this code cell

# laoding in the data
df = pd.read_csv("https://raw.githubusercontent.com/MIE223-2025/course-datasets/main/listings.csv")
print(df.shape)
df.head()
```

Creating `rating` column from `name` column by pulling all numbers after the `★` character.

```python
# Do NOT modify this code cell

# creating 'rating' column
star_char = "★"

df['rating'] = df['name'].apply(lambda x: re.findall(star_char + "(\\d+.?\\d*)", x)[0] if len(re.findall(star_char + "(\\d+.?\\d*)", x)) > 0 else np.nan)
df.head()
```

```python
print([col for col in df.select_dtypes(exclude = 'number').columns])

df['rating'] = df['rating'].astype(float)
print(df['rating'].dtype)
```

```python
print(df["rating"].describe()["75%"] - df["rating"].describe()["25%"])
```

```python
# Do NOT modify this code cell
print(f"df now has {df.shape[0]} rows")
print("\n")
```

```python
df['license'] = df['license'].fillna(df['license'].dropna().mode()[0])
df = df.dropna(subset = 'host_name')
(df.isnull().sum()/df.shape[0]).sort_values(ascending = False)
```

```python
# Do NOT modify this code cell

# Print the new number of rows in df
print(f"df now has {df.shape[0]} rows")
print("\n")
```

```python
print(f"Before imputing null values, 'price' column has: {df['price'].isnull().sum()} null values")


df['price'] = df['price'].fillna(df.groupby('neighbourhood')['price'].transform('mean'))


print(f"After imputing null values, 'price' column now has: {df['price'].isnull().sum()} null values")
```

```python
df['number_of_reviews_ltm'] = df['number_of_reviews_ltm'].apply(lambda x: "High" if pd.to_numeric(x, errors='coerce') > 70 else ("Low" if pd.to_numeric(x, errors='coerce') < 5 else "Medium"))
df['number_of_reviews_ltm'].value_counts()
```

```python
df.insert(loc = 0, column = "z-score_of_log_price", value = 0)

df_copy = df.copy()

df_copy['price'] = df_copy['price'].apply(lambda x: np.log(x))

df['z-score_of_log_price'] = df_copy['price'].apply(lambda x: (x - df_copy['price'].mean())/df_copy['price'].std())

df['z-score_of_log_price'].head(5)
```

```python
cols_to_output = ["id", "name", "host_id", "host_name", "neighbourhood", "room_type", "price", "number_of_reviews", "rating"]

zscore_threshold = 3

top_priced_listings = df[df['z-score_of_log_price'] < -3]

top_priced_listings = top_priced_listings[cols_to_output]

top_priced_listings = top_priced_listings.sort_values(by = 'price', ascending = False)

print(f"The shape of top_priced_listings is: {top_priced_listings.shape}")

top_priced_listings # Outputs top_priced_listings.
                    # Note, Google Colab may truncate this outputs to the first five and last five rows. If done, leave your output truncated by Google Colab.
```
